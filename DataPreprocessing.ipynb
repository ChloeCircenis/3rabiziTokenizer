```python
sp = spm.SentencePieceProcessor()
sp.load('spmthirty.model')
sp.encode(test.text[41], out_type=str)
```

```python
fertility_list = []

for sentence in test.text[:1000]:  # use a sample or full corpus
    words = sentence.strip().split()             # original words
    tokens = sp.encode(sentence, out_type=str)   # subword tokens
    if len(words) > 0:
        fertility = len(tokens) / len(words)
        fertility_list.append(fertility)

# corpus-level fertility
avg_fertility = sum(fertility_list) / len(fertility_list)
print("Average fertility:", avg_fertility)
```

```python
#get vocabulary size of full set
def get_vocab(data, vocab):
    for sample in data:
        text = sample.lower().split()
        for word in text:
            if word not in vocab:
                vocab.append(word)
    return vocab
```
